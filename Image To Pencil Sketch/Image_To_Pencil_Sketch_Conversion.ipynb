### **CONVERSION OF IMAGE TO PENCIL SKETCH**

**IMPORTING REQUIRED LIBRARIES**

```python
import cv2
from google.colab.patches import cv2_imshow
```

**READING THE ORIGINAL IMAGE**

```python
img = cv2.imread("/content/naruto.png")
cv2_imshow(img)
```

**CONVERTING TO GRAYSCALE IMAGE**

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

**GRAY SCALE IMAGE GIVES US BLACK & WHITE PIXELS WHICH IS USED FOR CREATING A PENCIL SKETCH.**

**INVERTING THE GRAY IMAGE**

```python
invert_img = cv2.bitwise_not(gray_img)
cv2_imshow(invert_img)
```

**IT WILL MAKE THE BRIGHTER REGIONS LIGHTER SO THAT WE CAN FIND EDGES TO CREATE PENCIL SKETCH**

**SMOOTHING & INVERTING OF IMAGE**

```python
smooth_img = cv2.GaussianBlur(invert_img, (21, 21),sigmaX=0 , sigmaY=0)
invert_smooth = 255-smooth_img
cv2_imshow(invert_smooth)
```

**THIS WILL CREATE THIN LINES FOR OUR SKECTH AND NOICE CAN BE REDUCED.**
**BY INVERTING WE CAN HIGHLIGHT THE BOLDEST EDGE.**

**OBTAINING THE PENCIL SKETCH**

```python
pencil_image = cv2.divide(gray_img,invert_smooth,scale=256)
cv2_imshow(pencil_image)
cv2.imwrite("naruto_pencil.jpg",pencil_image)
```
